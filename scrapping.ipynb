```python
url = 'https://www.imdb.com/chart/toptv'
```

```python
import re
import pandas as pd
import numpy as np
import itertools
from bs4 import BeautifulSoup
import requests
```

```python
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
soup
```

```python
tvtags = soup.find_all('a',attrs={'class': None})
```

```python
links_2 = []
for tag in tvtags:
    if tag.has_attr('href'):
        links_2.append(tag['href'])
links_2
```

```python
links = []
for e in links_2:
    if e.startswith('/title') & e.endswith('/'):
        links.append(e)

print(len(links))
print (links)
```

```python
links = list(dict.fromkeys(links))
print(len(links))

links[:10]
```

```python
base_url = 'https://www.imdb.com'
tv_links = [base_url + l + 'reviews' for l in links]
print (len(tv_links))
tv_links[:10]
```

```python
def getsoup(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    return soup
```

```python
tv_soups = [getsoup(link) for link in tv_links]
```

```python
def minmaxrev(x):
    minpos= x.index(min(x))
    maxpos= x.index(max(x))
    return minpos, maxpos
```

```python
def getreviews(soup):
    user_review_rat = [t.previous_element for t in soup.find_all('span', attrs={'class':'point-scale'})]
    neg_index, pos_index = minmaxrev(list(map(int, user_review_rat)))
    user_rev_list = soup.find_all('a', attrs={'class':'title'})
    n_rev_tag = user_rev_list[neg_index]
    p_rev_tag = user_rev_list[pos_index]
    n_rev_link = "https://www.imdb.com" + n_rev_tag['href']
    p_rev_link = "https://www.imdb.com" + p_rev_tag['href']
    return n_rev_link, p_rev_link
```

```python
review_list = [getreviews(tv_soup) for tv_soup in tv_soups]
```

```python
review_list = list(itertools.chain(*review_list))
```

```python
print(len(review_list))
review_list[:10]
```

```python
review_list = list(dict.fromkeys(review_list))
print(len(review_list))
review_list[:10]
```

```python
def getrevtxt(url):
    soup = getsoup(url)
    tag = soup.find('div', attrs={'class':'text show-more__control'})
    return tag.getText()
```

```python
rev_texts = [getrevtxt(url) for url in review_list]
```

```python
def gettitle(url):
    soup = getsoup(url)
    tag = soup.find('h1')
    return(list(tag.children)[1].getText())
```

```python
titles = [gettitle(url) for url in review_list]
```

```python
review_sentiment = np.array(['negative','positive'] * (len(review_list)//2))
```

```python
df = pd.DataFrame({'title': titles, 'user_review_link' : review_list, 'user_review': rev_texts, 'sentiment': review_sentiment})
```
